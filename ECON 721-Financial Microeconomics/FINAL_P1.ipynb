```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.optimize import fmin_cobyla
```

## (a)

```python
def optimal_portfolio(e0, e1, beta, theta):
    p0 = 1
    p1 = 0.5
    price_ratio = p0 / p1  # price_ratio = 2
    k = (beta * price_ratio) ** (1 / theta)
    numerator = p0 * e0 + p1 * e1
    denominator = p0 + p1 * k
    x0 = numerator / denominator
    x1 = x0 * k
    return x0, x1
```

## (b)

```python
e0 = 5
e1 = 20
beta = 0.5
theta = 5

x0_star, x1_star = optimal_portfolio(e0, e1, beta, theta)

def utility(x0, x1, beta, theta):
    return (x0 ** (1 - theta)) / (1 - theta) + beta * (x1 ** (1 - theta)) / (1 - theta)

U_star = utility(x0_star, x1_star, beta, theta)

x1_vals = np.linspace(0, e0 + 0.5 * e1 + 5, 200)
x0_budget = (e0 + 0.5 * e1) - 0.5 * x1_vals

x0_vals = np.linspace(max(x0_star * 0.5, 1e-6), x0_star * 1.5, 200)
x1_indiff = []

for x0 in x0_vals:
    term = ((1 - theta) * U_star - x0 ** (1 - theta)) / beta
    if term > 0:
        x1 = term ** (1 / (1 - theta))
        x1_indiff.append(x1)
    else:
        x1_indiff.append(np.nan)

x1_indiff = np.array(x1_indiff)
```

```python
plt.figure(figsize=(8,6))
plt.plot(x0_budget, x1_vals, label='Budget Line')
plt.plot(x0_vals, x1_indiff, label='Indifference Curve')
plt.scatter(e0, e1, color='red', label='Endowment Point')
plt.scatter(x0_star, x1_star, color='green', label='Optimal Choice')
plt.xlabel('$x_0$')
plt.ylabel('$x_1$')
plt.title('Optimal Choice with Endowment (5,20), beta=0.5, theta=5')
plt.legend()
plt.grid(True)
plt.show()
```
